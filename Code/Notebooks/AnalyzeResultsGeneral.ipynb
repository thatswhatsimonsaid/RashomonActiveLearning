# Save

```python
### Save ###
time_table = AnalysisResultsOutput.get("TimeTable")
time_table_path = os.path.join(output_folder, "TimeTable.csv")
time_table.to_csv(time_table_path)

trace_plot_mean = AnalysisResultsOutput.get("TracePlotMean")
trace_plot_path = os.path.join(output_folder, "TracePlot_Mean.png")
trace_plot_mean.savefig(trace_plot_path, bbox_inches='tight', dpi=300)

refit_plot = AnalysisResultsOutput.get("RefitFrequencyPlot")
refit_plot_path = os.path.join(output_folder, "RefitFrequencyPlot.png")
refit_plot.savefig(refit_plot_path, bbox_inches='tight', dpi=300)

tree_plot_unreal = AnalysisResultsOutput.get("TreePlot_UNREAL")
tree_plot_unreal_path = os.path.join(output_folder, "TreePlot_UNREAL.png")
tree_plot_unreal.savefig(tree_plot_unreal_path, bbox_inches='tight', dpi=300)

tree_plot_dureal = AnalysisResultsOutput.get("TreePlot_DUREAL")
tree_plot_dureal_path = os.path.join(output_folder, "TreePlot_DUREAL.png")
tree_plot_dureal.savefig(tree_plot_dureal_path, bbox_inches='tight', dpi=300)
```
